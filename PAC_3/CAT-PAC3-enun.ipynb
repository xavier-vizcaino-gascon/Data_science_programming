Estudieu la complexitat algorísmica, considerant el pitjor cas, en relació amb el temps d'execució del següent fragment de codi. Podem utilitzar la notació $\Theta$ per referir-nos a la complexitat del codi analitzat? En cas afirmatiu, mostreu gràficament l'orde de creixement de la funció juntament amb una funció que faci de límit superior i una altra que faci de límit inferior.

**Nota**: Recordeu repassar la secció 2.2.3 Notacions i definicions formals del Notebook de teoria.

<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">NM</span> **(2 punts)**

```python
def findXYZ(x):
    solutions = []

    for i in range(x):
        for j in range(x):
            for k in range(x):
                if 3*i + 9*j + 8*k == 79:
                    solutions.append((i, j, k))

    return solutions
```

**Resposta**

Recorrent als passos que cal executar per l'algorisme implementat a la funció findXYZ tenim:

- Creació de la llista (1)
- Per cada element de la llista exterior (n)
    - Recuperar i
- Per cada element de la llista intermitja (n (i) * n (j) = n²)
    - Recuperar j
- Per cada element de la llista interior (n (i) * n (j) * n (k) = n³)
    - Recuperar k
    - Comparar si es compleix la equació
    - Si es compleix afegir la tupla a la llista.

Per tant la complexitat algorísmica és:$$t_e(n) = 1 + n + n² + 2n³ + x, \text{  on  }x\in[0,n³]$$ I pel pitjor cas passa a ser: $$t_e(n) = 1 + n + n² + 3n³$$

Addicionalment podem dir que $1 + n + n² + 3n³ \in \Theta(n³)$
Ja que, a tall d'exemple, per a $c_1=2$ i $c_2=4$, $n_0=1.84$, es compleixen les condicions següents:

$$\Theta(g(n)) = \{ f(n) : 
    \exists c_1, c_2 \text{ i } n_0 > 0
    \text{ tals que  }
    0 \leq c_1g(n) \leq f(n) \leq c_2g(n),
    \forall n \ge n_0 \}$$

Es demostra gràficament amb el següent codi:

```python
# codi reciclat dels apunts
import matplotlib.pyplot as plt
import numpy as np


def plot_funs_theta(xs, n_val=True, sup_only=False):
    """
    Plot the function f(n) = 1+n+n²+3n³ together with:
    - lower bound c1 * g(n) = c1*n³
    - upper bound c2 * g(n) = c2*n³.
    """

    # Inicialitzem variables
    legend, c1, c2, n0 = [], 2, 4, 1.84

    # Generem arrays i imputem valor a llegenda
    ys_f, leg_f = [1+x+x**2+3*(x**3) for x in xs], "$f(n)$"
    ys_low, leg_low = [c1*(x**3) for x in xs], "$c1*g(n)$"
    ys_high, leg_high = [c2*(x**3) for x in xs], "$c2*g(n)$"

    # Creem el plot
    fig = plt.figure()
    plt.plot(xs, ys_f, '-', color='tab:orange')
    legend.append(leg_f)
    plt.plot(xs, ys_high, ':', color='tab:green')
    legend.append(leg_high)

    # Afegim variable si hem de graficar ambdos limits
    if not sup_only:
        plt.plot(xs, ys_low, ':', color='tab:blue')
        legend.append(leg_low)

    # Afegim linia amb valor n0
    if n_val:
        plt.plot([n0, n0], [min(ys_high), max(ys_high)], '--', color='tab:gray')

    plt.legend(legend)
    plt.xlabel('$n$')
    plt.ylabel('$f(n)$')
    plt.title('Function growth')
    plt.show()


plot_funs_theta(np.arange(0, 100, 1))
plot_funs_theta(np.arange(1.2, 2.5, 0.1))

# REFERENCIES
# https://pynative.com/python-range-for-float-numbers/
# funció per a graficar: plot_funs_theta -> Document 4-CAT-Optimització_complexitat_i_profiling
```

```python
# codi propi reutilitzable


def my_plotter(x: list, y1: list, y2: list = [], y3: list = [],
               y1_legend: str = "", y2_legend: str = "",
               y3_legend: str = "", x_label: str = "",
               y_label: str = "", plot_title: str = ""):
    """
    Plot values of y1, y2, y3 against values of x.
    y2 & y3 parameters are optional.
    """
    # Inicialitzem llegenda
    legend: list = []

    # Generem el plot
    fig = plt.figure()
    plt.plot(x, y1, '-', color='tab:orange')
    legend.append(y1_legend)

    # Si tenim dades en la variable y2 la plotejem
    if len(y2) != 0:
        plt.plot(x, y2, ':', color='tab:green')
        legend.append(y2_legend)

    # Si tenim dades en la variable y3 la plotejem
    if len(y3) != 0:
        plt.plot(x, y3, ':', color='tab:blue')
        legend.append(y3_legend)

    # Afegim llegenda, labels als eixos i títol
    plt.legend(legend)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(plot_title)
    plt.show()


# Generem arrays de dades per a x i per a f(x)
x_values = np.arange(0, 100, 1)
y_values: list = [1+x+x**2+3*(x**3) for x in x_values]

# Inicialitzem variables
c1, c2, n0 = 2, 4, 1.84
# Generem arrays de dades per a g(x)
ylow_values: list = [c1*(x**3) for x in x_values]
yhi_values: list = [c2*(x**3) for x in x_values]

# Plotejem dades
my_plotter(x_values, y_values, ylow_values, yhi_values, "$f(n)$",
           "$c1*g(n)$", "$c2*g(n)$", "$n$", "$f(n)$",
           "Function growth")
```

```python
def list_unique(array):
    """Returns unique elements in the array.

    Parameters:
        array (list): input array for searching unique elements.

    Returns:
        (list): list of unique elements.


    If an element repeats itself it will be ignored from
    the final result. For instance, list_unique([1, 2, 2, 4])
    will return [1, 4].
    """

    unique = []
    i = 0

    while i < len(array):
        j = 0
        count = 0
        while j < len(array):
            if i != j and array[j] != array[i]:
                count = count + 1

            j = j + 1

        # If the element is different from all the others,
        # we consider it to be unique.
        if count == len(array) - 1:
            unique.append(array[i])

        i = i + 1

    return unique
```

**Resposta**

```python
# Test_inicial

my_array = [1, 2, 2, 3, 3, 4, 4, 5, 5]
%lprun -f list_unique list_unique(my_array)
```

```python
# Alternativa_1


def list_unique_1(array):
    """Returns unique elements in the array using list count() method
    within a for loop.

    Parameters:
        array (list): input array for searching unique elements.

    Returns:
        (list): list of unique elements.


    If an element repeats itself it will be ignored from
    the final result. For instance, list_unique([1, 2, 2, 4])
    will return [1, 4].
    """

    unique: list = []
    for i in array:
        if (array.count(i) == 1):
            unique.append(i)

    return unique
```

```python
# Alternativa_2


def list_unique_2(array):
    """Returns unique elements in the array using count() method
    within list comprehension.

    Parameters:
        array (list): input array for searching unique elements.

    Returns:
        (list): list of unique elements.


    If an element repeats itself it will be ignored from
    the final result. For instance, list_unique([1, 2, 2, 4])
    will return [1, 4].
    """
    return [i for i in array if array.count(i) == 1]
```

<span style="font-family: Courier New; background-color: #ffcc5c; color: #000000; padding: 3px; ">EG</span> **(1.5 punts)**

```python
import random


def count_appearence(iters, n):
    """Given a number n, generates a random list of 'iters'
     number of elements between [-7*n,7*n] and returns the indexes at which
     our input n appears in the random list.

     Parameters:
         iters (int): integer, number of random elements to generate
         n (int): integer, number to count appearences

     Returns:
         (list): list if indexes of the generated random list
                 where `n` appears as value.

     For instance, if the generated list is [-7, 5, 3, 1, 4, 1, 7] and `n = 1`,
     then the result would be [3, 5] since `1` is present at indexes 3 and 5
     respectively of the array.
    """

    c = 0
    L = []
    Idx = []
    for i in range(iters):
        L.append(random.randint(-7*n, 7*n))

    for j in range(len(L)):
        if L[j] == n:
            Idx.append(j)

    return Idx
```

```python
# PUNT 1
import os
import json
import matplotlib.pyplot as plt
import datetime

# Funció per a comprovar arxiu i carregar dades
def load_json(file: str) -> dict:
    '''
    La funció rep un path a un arxiu json i en carrega les dades
    generant un diccionari
    '''
    # Comprovem que l'arxiu existeixi
    if not os.path.isfile(file):
        raise TypeError("File does not exist")

    # Obrim arxiu i carreguem dades
    with open(file, "r") as f:
        data = json.load(f)
    return data

# Funció per extreure el nombre d'aproximacions totals i perilloses
# per a cada data dins les dades
def get_approach_number(data: dict) -> list:
    '''
    La funció rep un diccionari com a dades d'entrada i retorna una llista
    de tuples.
    En cada tupla trobem:
    * La data a la que fan referència les dades
    * El nombre d'aproximacions totals per a la data indicada
    * El nombre d'aproximacions considerades perilloses
    '''
    # Inicialitzem llista de sortida
    L_out: list = []
    # Iterem sobre tots els valors de les dades per obtenir la informació
    # i generar les tuples
    for key, value in data.items():
        approaches: int = len(value)
        hazardous: int = len([index for index in value if
                         index['is_potentially_hazardous_asteroid']])
        L_out.append((key, approaches, hazardous))
    return L_out

# Funció per a pre-processar les dades abans de graficar
def manage_data(data: list) -> dict:
    '''
    La funció rep una llista de tuples, genera una nova variable i retorna
    el resultat en un diccionari on les claus son les diferents variables
    i els valors son llistes ordenades d'acord a la data
    '''
    # Generem llistes planes a partir de les tuples
    date_list: list = [i[0] for i in data]
    app_list: list = [i[1] for i in data]
    haz_list: list = [i[2] for i in data]

    # Creem variable addicional per a les aproxiacions no perilloses
    no_haz_list: list = [x-y for x, y in zip(app_list, haz_list)]

    # Ordenem els valors de totes les llistes agafant com a referència la data
    zipped_lists: list = list(zip(date_list, app_list, haz_list, no_haz_list))
    ordered: list = sorted(zipped_lists, key=lambda x: x[0])
    unzipped: list = list(zip(*ordered))

    return {"Date": list(unzipped[0]), "Approaches": list(unzipped[1]),
            "Hazardous": list(unzipped[2]), "Not_hazardous": list(unzipped[3])}


# Funció per a crear un gràfic dedicat
def create_chart(x_data: list, y_data: dict):
    fig, ax = plt.subplots()
    # Convertim dades de dates a datetime
    dates: list = [datetime.datetime.strptime(x, "%Y-%m-%d") for x in x_data]

    ax.stackplot(dates, y_data.values(),
                 labels=y_data.keys(), alpha=0.8)
    ax.legend(loc='upper left')
    ax.set_title('Approaches')
    ax.set_xlabel('Date')
    ax.tick_params(axis='x', labelrotation=45)
    ax.set_ylabel('Number')
    plt.show()


# Definim arxiu d'entrada
p: str = "data/close-approaches.json"

# Generem diccionari
data: dict = load_json(p)

# Obtenim aproximacions
app: list = get_approach_number(data)

# Processem dades
info: dict = manage_data(app)

# Creem gràfic específic
create_chart(info["Date"], {"Potentially_hazardous": info["Hazardous"],
                            "Not_hazardous": info["Not_hazardous"]})

# Canviem format a les dades x a datetime
dates: list = [datetime.datetime.strptime(x, "%Y-%m-%d") for x in info["Date"]]

# Creem gràfic reciclant codi del exercici 2
my_plotter(x=dates, y1=info["Approaches"], y2=info["Hazardous"],
           y1_legend="Total_approaches", y2_legend="Potentially_hazardous",
           x_label="Date", y_label="Number",
           plot_title="Approaches")


# REFERENCES
# https://matplotlib.org/stable/gallery/lines_bars_and_markers/stackplot_demo.html#sphx-glr-gallery-lines-bars-and-markers-stackplot-demo-py
# https://www.geeksforgeeks.org/python-ways-to-sort-a-zipped-list-by-values/
```

```python
# PUNT 3a/b- MULTI (segons apunts)
# GENERALS
from threading import Thread, current_thread
from multiprocessing import JoinableQueue, Queue


def search_write(m_index: int, neo_to_process_q, API_k: str, f_path: str):
    '''
    La funció repo com a paràmetres:
    * L'index de multi (thread o procés)
    * Una cua amb els neos a processar (tasques)
    * LA clau API
    * El path al directori
    
    De manera anàloga la funció genera un arxiu csv per a cada neo,
    però en aquest cas està preparada per a funcionar en mode
    multi-threading i/o multi-processing
    '''
    logging.info("[Multi_{}]\tStarted".format(m_index))

    # Obtenir la primera tasca
    neo_to_process = neo_to_process_q.get()

    # Mentre hi hagi tasques pendents
    while neo_to_process:
        logging.info("[Multi_{}]\tStarting to work on {}".
                     format(m_index, neo_to_process))
        # Obtenim la informació de la API
        temp: dict = search_by_neo_id(API_k, neo_to_process)
        # Generem el file path
        neo_name: str = temp['name']
        file_path = os.path.join(f_path, neo_to_process+".csv")
        # Obrim l'arxiu en mode escriptura
        with open(file_path, 'w', newline='') as csvfile:
            writer = csv.writer(csvfile, delimiter=",")
            # Creem la capcelera
            writer.writerow(["neo_name", "close_approach_date",
                             "velocity", "lunar_distance"])
            # Afegim les dades
            for elements in temp['close_approach_data']:
                date: str = elements['close_approach_date']
                velocity: str = elements['relative_velocity']['kilometers_per_hour']
                distance: str = elements['miss_distance']['lunar']
                writer.writerow([neo_name, date, velocity, distance])

        logging.info("[Multi_{}]\t{} ending".format(m_index, neo_to_process))
        # Informem de tasca acabada
        neo_to_process_q.task_done()
        # Obtenim nova tasca
        neo_to_process = neo_to_process_q.get()

    logging.info("[Multi_{}]\tEnding".format(m_index))
    # Informem de totes les tasques acabades
    neo_to_process_q.task_done()
    logging.info("[Multi_{}]\tProcess End".format(m_index))
```

```python
# PUNT 3a MULTI-THREAD (segons apunts)


def data_process_MT(nl: list, num_threads: int, API: str, path: str):
    '''
    La funció governa la creació d'arxius csv en mode multi-threading.
    Els paràmetres d'entrada son:
    * Una llista amb els neos
    * El nombre màxim de threads
    * La clau API
    * El path al directori on crear els arxius csv
    '''
    # Generem el directori si no existeix
    try:
        os.makedirs(path)
    except OSError:
        pass

    # Creem cua de tasques a realitzar
    neo_to_process_q = JoinableQueue()

    # Afegim neos a la cua de tasques
    for neo in nl:
        neo_to_process_q.put(neo)

    # Afegim indicadors de final
    for _ in range(num_threads):
        neo_to_process_q.put(None)

    logging.info("[Main]\t\tMulti-threading initialization")
    threads: list = []
    # Iniciem processos passant cua com a paràmetre
    for i in range(num_threads):
        thread = Thread(target=search_write,
                        args=(i, neo_to_process_q, API, path))
        threads.append(thread)
        thread.start()

    # Esperem que els tres *threads* finalitzin
    for i, thread in enumerate(threads):
        logging.info("[Main]\t\tWaiting to join thread {}".format(i))
        thread.join()
        logging.info("[Main]\t\tThread joined")
    logging.info("[Main]\t\tROUTINE END")


data_process_MT(neos_list, 6, myAPI_key, folder_path)
```

```python
# PUNT 3b - MULTI-PROCESS (segons apunts)
from multiprocessing import Process


def data_process_MP(nl: list, num_processes: int, API: str, path: str):
    '''
    La funció governa la creació d'arxius csv en mode multi-processing.
    Els paràmetres d'entrada son:
    * Una llista amb els neos
    * El nombre màxim de threads
    * La clau API
    * El path al directori on crear els arxius csv
    '''
    # Generem el directori si no existeix
    try:
        os.makedirs(path)
    except OSError:
        pass

    # Creem cua de tasques a realitzar
    neo_to_process_q = JoinableQueue()

    # Afegim neos a la cua de tasques
    for neo in nl:
        neo_to_process_q.put(neo)

    # Afegim indicadors de final
    for _ in range(num_processes):
        neo_to_process_q.put(None)

    logging.info("[Main]\t\tMulti-process initialization")
    # Iniciem processos passant cua com a paràmetre
    for i in range(num_processes):
        process = Process(target=search_write,
                          args=(i, neo_to_process_q, API, path))
        process.start()

    # Esperem que es completin les tasques
    logging.info("[Main]\t\tWaiting to join processes")
    neo_to_process_q.join()
    logging.info("[Main]\t\tProcesses joined")
    logging.info("[Main]\t\tROUTINE END")


data_process_MP(neos_list, 6, myAPI_key, folder_path)
```

```python
# PUNT 3a - MULTI-THREAD (fent us de la classe ThreadPoolExecutor)


def data_process_MT_Pool(nl: list, num_threads: int, API: str, path: str):
    '''
    La funció gestiona lexecució de la funció search_write_Pool en mode
    multi-threading a través de ThreadPoolExecutor
    
    La funció rep com a paràmetres d'entrada:
    * Llista de neos a processar
    * Nombre màxim de threads
    * Clau API
    * Path al directori on crear els csv
    
    La funció crida la funcio search_write_Pool en mode multithread per tots
    els neos de la llista'''
    # Creem el directori neo si no existeix
    try:
        os.makedirs(path)
    except OSError:
        pass

    # Creem el ThreadPoolExecutor
    with concurrent.futures.ThreadPoolExecutor(max_workers=num_threads) as executor:
        # Definim funció d'execucio, paràmetres a passar i llista d'execució
        future_to_search_write = {executor.submit(search_write_Pool, neo, API, path): neo for neo in nl}
        for future in concurrent.futures.as_completed(future_to_search_write):
            neo = future_to_search_write[future]
            # Comprovem que hem obtingut resultat vàlid, sino generem excepció
            try:
                data = future.result()
            except Exception as exc:
                logging.info("[Main]\t{} generated an exception {}".
                             format(neo, exc))
            else:
                logging.info("[Main]\t{} done".format(neo))


data_process_MT_Pool(neos_list, 6, myAPI_key, folder_path)
```

```python
# PUNT 3b - MULTI-PROCESS (fent us de la classe ThreadPoolExecutor)


def data_process_MP_Pool(nl: list, num_processes: int, API: str, path: str):
    '''
    La funció gestiona lexecució de la funció search_write_Pool en mode
    multi-processing a través de ProcessPoolExecutor
    
    La funció rep com a paràmetres d'entrada:
    * Llista de neos a processar
    * Nombre màxim de threads
    * Clau API
    * Path al directori on crear els csv
    
    La funció crida la funcio search_write_Pool en mode multitprocess per tots
    els neos de la llista'''
    # Creem el directori neo si no existeix
    try:
        os.makedirs(path)
    except OSError:
        pass
    # Creem el ProcessPoolExecutor
    with concurrent.futures.ProcessPoolExecutor(max_workers=num_processes) as executor:
        # Definim funció d'execucio, paràmetres a passar i llista d'execució
        future_to_search_write = {executor.submit(search_write_Pool, neo, API, path): neo for neo in nl}
        for future in concurrent.futures.as_completed(future_to_search_write):
            neo = future_to_search_write[future]
            # Comprovem que hem obtingut resultat vàlid, sino generem excepció
            try:
                data = future.result()
            except Exception as exc:
                logging.info("[Main]\t{} generated an exception {}".
                             format(neo, exc))
            else:
                logging.info("[Main]\t{} done".format(neo))


data_process_MP_Pool(neos_list, 6, myAPI_key, folder_path)
```

```python
# PUNT 5
from collections import defaultdict

neo_id: str = "3989253"
folder_path: str = "data/neo"


def get_data(n_id: str, folder: str) -> dict:
    '''
    La funció rep com a paràmetres d'entrada un identificador
    de neo i un directori on buscar l'arxiu csv amb el nom del neo
    
    La funció retorna un diccionari amb totes les dades de:
    * neo_id
    * data
    * velocitat
    * distància
    
    Per al neo_id proporcionat
    '''
    # Generem el filepath
    file_path: str = os.path.join(folder, n_id+".csv")

    # Comprovem que l'arxiu existeixi
    if not os.path.isfile(file_path):
        raise TypeError("File does not exist")

    # Inicialitzem llista i defaultdict
    read_list: list = []
    out: dict = defaultdict(list)

    # Obrim arxiu csv en format lectura
    with open(file_path, 'r', newline='') as csvfile:
        reader = csv.DictReader(csvfile, delimiter=",")
        # Obtenim files
        for row in reader:
            read_list.append(row)
    # Generem diccionari de sortida
    for records in read_list:
        for k, v in records.items():
            if k == "close_approach_date":
                out[k].append(datetime.datetime.strptime(v, "%Y-%m-%d"))
            elif k == "velocity":
                out[k].append(float(v))
            elif k == "lunar_distance":
                out[k].append(float(v))
            else:
                out[k].append(v)
    return out


info: dict = get_data(neo_id, folder_path)

# codi reciclat d'exercicis prèvis
my_plotter(x=info["close_approach_date"], y1=info["velocity"],
           y2=info["lunar_distance"], y1_legend="velocity",
           y2_legend="Lunar_distance", x_label="Date",
           y_label="Number", plot_title="Approaches")
```

```python
# nou codi amb més opcions gràfiques
from scipy import interpolate


def interpolate_data(x_list: list, y_list: list) -> tuple:
    '''
    La funció rep com a paràmetres d'entrada
    una llista amb valors de x i una llista amb valors de
    y
    La funció retorna una tupla amb els valors de x i de y
    amb més definició (10.000 punts) interpolats amb el mètode
    "nearest"
    '''
    # Transformem valors x de datetime a timestamp
    x_values: list = [datetime.datetime.timestamp(x) for x in x_list]

    # Generem nou array amb 10.000 valors entre els valors màxims
    # i minims de la variable x
    x: list = np.array(x_values)
    x_new: list = np.linspace(x.min(), x.max(), 10000)

    # Definim la funció d'interpolació par a la variable y
    f: list = interpolate.interp1d(x_values, y_list, kind="nearest")
    y1_new: list = f(x_new)

    # Convertim la variable x (amb més definició) de timestamp 
    # a datetime 
    x_n: list = [datetime.datetime.fromtimestamp(x) for x in x_new]

    return (x_n, y1_new)


velo_tuple: tuple = interpolate_data(info["close_approach_date"],
                              info["velocity"])
luna_tuple: tuple = interpolate_data(info["close_approach_date"],
                              info["lunar_distance"])

# Generem gràfic
fig, axs = plt.subplots(2, 1, sharex=True)
axs[0].plot(velo_tuple[0], velo_tuple[1], color="dimgray")
axs[1].plot(luna_tuple[0], luna_tuple[1], color="darkgray")
for i in 0, 1:
    axs[i].set_xlabel("year")
    axs[i].grid(True)
axs[0].set_ylabel("velocity [km/h]")
axs[1].set_ylabel("lunar_distance")

fig.tight_layout()
axs[0].set_title("NEO_id {}".format(neo_id))
plt.show()

# REFEERENCES
# https://stackoverflow.com/questions/5283649/plot-smooth-line-with-pyplot
```

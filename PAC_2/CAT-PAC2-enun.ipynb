# Exercici 2

Una agència de premsa està elaborant un reportatge sobre l'opinió de la població respecte al canvi climàtic i ha sol·licitat la nostra col·laboració per identificar quins són els temes més actuals en aquest àmbit. Per ajudar-los, hem de crear una funció que extregui el contingut dels hashtags que comencen per la paraula "climate" (sense tenir en compte si està escrit en majúscules o minúscules) d'una col·lecció de tweets.


La funció rebrà com a *input* la ruta que conté el fitxer amb els tweets en format *.csv*. L'*output* serà una tupla amb el format `(nombre_tweets, llista_hashtags)`, on  
- `nombre_tweets` indica el nombre de tweets que contenen hashtags.
- `llista_hashtags` és la llista dels hashtags **sense repeticions**

El fitxer d'entrada que farem servir és `twitter_sentiment_data.csv`, que està ubicat a la carpeta `data` i ha estat extret de https://www.kaggle.com/edqian/twitter-climate-change-sentiment-dataset?select=twitter_sentiment_data.csv.


Aquest fitxer està compost per una capçalera seguida per un nombre indeterminat de línies amb el format
```
sentiment, message, tweetid
```

De manera que cada línia del fitxer després de la capçalera conté un tweet.

S'hi inclou, a més, una cel·la extra amb el codi que heu d'utilitzar per comprovar que la funció s'executa correctament.

**Important:** Feu servir els principis de **programació funcional** que hem vist al Notebook de teoria per resoldre aquest exercici i **expressions regulars** per detectar els hashtags.

<span style="font-family: Courier New; background-color: #ffcc5c; color: #000000; padding: 3px; ">EG</span>  **(2 punts)**

### Resposta 2

````python
# Resposta 2
import re


def find_hashtags(path: str) -> tuple:
    '''
    Retorna una tupla amb:
    - El nombre de tweets on apareix com a mínim 1 hashtag
    que comenci amb #climate.
    - El llistat de hashtags que comencen amb #climate.

    Aquesta funció accepta un sol argument de tipus string
    que defineix la ruta de l'arxiu origen de dades. L'arxiu,
    com a mínim ha de constar d'una columna anomenada "message"
    on es trobin els tweets.

    La funció no té en compte majuscules o mínuscules per a la
    detecció de hashtags.

    La funció si té en compte que en un mateix tweet hi pot haver
    més d'un hashtag objectiu, i consequentment els detecta tots.
    '''

    # Carreguem dades
    dades: pd.DataFrame = carregar(path)

    # Llista de tweets que contenen Hashtags que comencen amb #climate
    twt_list: list = [twt for twt in list(dades["message"]) if
                      re.search(r"#climate\w+", twt, re.IGNORECASE)]

    # Nombre de tweets
    num_twt: int = len(twt_list)

    # Extreiem hashtags de la llista de tweets
    climates: list = [re.findall(r"#climate\w+", twt, re.IGNORECASE) for
                      twt in twt_list]

    # Aplanem la llista i en fem un grup (valors únics)
    flat_climates: list = [i for sub in climates for i in sub]
    hashtags: list = sorted(list(set(flat_climates)))

    return (num_twt, hashtags)

# REFERENCIES
# https://docs.python.org/3/library/re.html#match-objects
# https://stackoverflow.com/questions/952914/how-to-make-a-flat-list-out-of-a-list-of-lists
````

L'Agència Estatal de Meteorologia va a elaborar el seu informe triennal sobre l'evolució de la contaminació en el període 2018-2020. Per això, ens han facilitat l'arxiu `Pollution.csv`, que està ubicat a la carpeta `data`. Les dades han estat obtingudes de https://www.aemet.es/es/datos_abiertos/estadisticas/composicion_quimica_atmosfera. Aquest fitxer conté els registres de concentració de diferents molècules en micrograms per metre cúbic d'aire obtinguts a diferents estacions meteorològiques. Les dades es registren sempre el darrer dia de cada mes i indiquen concentracions mensuals.

La nostra missió consisteix a crear dues funcions. En primer lloc, `create_dictionary` que rebrà com a *input* el path al fitxer de dades, és a dir, `data/Pollution.csv` i una data en format string separada por guions (per exemple, \"15-03-2020\"). La funció `save_pickle` ha de produir un **diccionari** guardat a un **pickle** de manera que per a cada molècula es guardin dues llistes, una amb les dates de registre consultades i altra amb les concentracions mensuals associades a aquestes dates.

Fent una anàlisi preliminar de l'arxiu, observem que hi ha dades de quatre molècules diferents (SO2, O3, NO i NO2) enregistrades des de 13 estacions. Per tant, haurem d'agrupar per *Molecule* i calcular la mitjana de concentracions a les diferents estacions. Finalment, com que aquestes dades seran analitzades como a sèries temporals, crearem un diccionari amb la següent estructura:

```
{
"Molecule_1": {"registration date": [31/01/18, 28/02/18,...], "monthly concentration (ug/m3)": [0, 0,...]},
"Molecule_2": {"registration date": [31/01/18, 28/02/18,...], "monthly concentration (ug/m3)": [0, 0,...]},
"Molecule_3": {"registration date": [31/01/18, 28/02/18,...], "monthly concentration (ug/m3)": [0, 0,...]},
"Molecule_4": {"registration date": [31/01/18, 28/02/18,...], "monthly concentration (ug/m3)": [0, 0,...]},
}
```
On "Molecule_N" ha de ser substituït pel nom de la molècula. És important tenir en compte que les dates a l'input de la funció són strings separats per guions, però al diccionari seran variables datetime separades per barres.

**Nota:** cal tenir en compte que les dades estan en format dd/mm/yyyy.

Una vegada construït el diccionari, el guardarem a un pickle que anomenarem `pollution_data.pkl`. S'hi inclou, a més, una cel·la extra amb el codi que heu d'utilitzar per comprovar que la funció s'executa correctament.


<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">SM</span> **(2 punts)**

### Resposta 3

````python
# Resposta 3
import datetime
import pickle


def create_dictionary(ruta: str, limit_date: str) -> dict:
    '''
    Retorna un diccionari amb:
    - Les molecules d'analisi (str).
    - Un llistat de les dates d'adquisició (datetime).
    - Un llistat amb les concentracions per cada data
    d'adquisició (float).

    Aquesta funció accepta dos arguments posicionals:
    - Ruta a les dades d'origen.
    - Data limit per a les adquisicions.

    La funció realitza el promitg per a cada molècula
    i data, de les diferents estacions d'adquisició.

    La funció limita les dades de sortida, retornant
    sempre les dades des de l'origen de les adquisicions
    fins a la data especificada com a data límit.
    '''
    # Funció per a convertir les dates del dataframe a objectes datetime
    def x_date(dia: str) -> datetime:
        '''
        La funció retorna una data en format datetime.

        La funció accepta un sol argument (str)
        que especifica una data en format dd/mm/YYYY.

        La funció converteix l'entrada i la retorna com
        a objecte datetime.
        '''
        # Convertim string a datetime
        datahora: datetime = datetime.datetime.strptime(
            dia, "%d/%m/%Y")
        return datahora

    # Funció per a convertir les dates del dataframe
    # al format sol·licitat passant per objecte datetime
    def x2_date(dia: str) -> str:
        '''
        Retorna una data tipus str en format dd/mm/YYYY.

        La funció accepta un sol argument (str)
        que especifica una data en format dd/mm/YYY.

        La funció converteix la data dentrada a un
        objecte datetime i posteriorment la torna
        a convertir a str en format dd/mm/YYYY.
        '''
        return x_date(dia).strftime("%d/%m/%Y")

    # Carregar dades
    dades: pd.DataFrame = carregar(ruta)

    # Agrupar dades per "molecule" fent la mitja de les diferents
    # lectures d'estacions
    dades_avg: pd.DataFrame = dades.groupby(by="Molecule").mean()

    # Convertir limit_date a datetime
    limit_date: datetime = datetime.datetime.strptime(
            limit_date, "%d-%m-%Y")

    # Generar llista de dates del dataframe limitada per limit_date
    date_list: list = [dia for dia in dades_avg.columns if
                       x_date(dia) <= limit_date]

    # Creació del diccionari
    dict_1: dict = {mol: {"registration date": [x2_date(dia) for
                                                dia in date_list],
                          "monthly concentration (ug/m3)": [round(float(dades_avg.at[mol, dia]), 4) for
                                                            dia in date_list]} for
                    mol in dades_avg.index}
    return dict_1


def save_pickle(path: str, dia: str):
    '''
    La funció executa la funció create_dictionay i
    crea un pickle amb el resultat.

    Aquesta funció accepta 2 arguments que són
    transmesos a la funció create_dictionary.

    La funció guarda un pickle en la ruta data/
    amb el nom pollution_data.pkl.
    '''
    # Creem diccionari
    dict_1: dict = create_dictionary(path, dia)

    # Definim arxiu de sortida
    p: str = "data/pollution_data.pkl"

    # Obrim arxiu i generem pickle
    with open(p, "wb") as f:
        pickle.dump(dict_1, f)

# REFERENCIES
# https://pandas.pydata.org/Pandas_Cheat_Sheet.pdf
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.groupby.html
````

Després de consolidar la seva posició al mercat europeu, una empresa de transport vol expandir el seu negoci als Estats Units. Per això ens ha encarregat que definim una o diverses funcions per crear una estructura ordenada de fitxers a partir del fitxer d'entrada amb els codis postals de cada ciutat.

El fitxer d'entrada que farem servir és `US_zipcodes.txt`, que està ubicat a la carpeta `data` i ha estat extret de https://github.com/midwire/free_zipcode_data

Aquest fitxer no té cap encapçalament i conté un nombre indeterminat de línies amb el següent format:

state, city, zipcode

on
- `state` és un codi de dos caràcteres que indica l'estat.
- `city` és el nom de la ciutat.
- `zipcode` és el codi postal.

La sortida esperada és una estructura de directoris i fitxers com la descrita a continuació:

```

    <state1>
        <city1>
            zipcode.txt

Dintre de cada fitxer `zipcode.txt` agruparem les dades del fitxer original per ciutat (cada ciutat pot contenir diversos codis postals) seguint el següent format:
```
Number of zipcodes: number_zipcodes
Zipcode #1: zipcode
Zipcode #2: zipcode

Zipcode #N: zipcode

On `number_zipcodes` és el nombre de codis postals diferents que té la ciutat i `zipcode` és el codi postal. El directori a partir del qual es genera aquesta estructura haurà de ser un paràmetre de la funció.

Un cop creada l'estructura, necessitem una altra funció que mostri per pantalla la ruta de les carpetes corresponents a cada estat i el nombre de subcarpetes (ciutats) que conté cada estat de la següent manera:

### Resposta 4

````python
# Resposta 4
import os
import subprocess as sp


def create_file_system(in_file: str, tgt_dir: str):
    '''
    La funció genera una estructura de carpetes i
    arxius amb la informació dels codis postals de USA.

    Aquesta funció accepta dos arguments:
    - Ruta en format string a arxiu de text pla, sense
    capçalera amb les dades dels estats, ciutats i
    codis postals d'estats units.
    - Ruta on generar l'estructura de carpetes

    Amb la informació dentrada, la funció genera,
    en el directori objectiu una carpeta per a cada estat.
    Dins de cada carpeta amb el nom de l'estat, la funció
    genera una carpeta per a cada ciutat.
    Dins de cada carpeta amb el nom de la ciutat, la funció
    crea un arxiu de text amb:
    - El nombre de codis postals diferents que hi ha
    en aquella ciutat.
    - Un llistat dels codis postals.
    '''

    def create_folders(path: str):
        '''
        La funció genera una carpeta en la ruta indicada
        '''
        # Creem carpeta si no existeix
        try:
            os.makedirs(path)
        except OSError:
            pass  # Si ja existeix s'activarà la excepció.

    def create_text_file(f: str, ds: pd.Series):
        '''
        La funció genera un arxiu de text en la ruta
        proprcionada per f amb el numero d'items en la
        serie d'entrada + un llistat dels items.
        '''
        with open(f, "w") as out:

            # Esciure número de codis
            out.write("Number of zipcodes: {}\n".format(ds.size))

            # Loop a través de tots els codis de la ciutat escrivint
            for value in enumerate(ds.iteritems()):
                out.write("Zipcode #{}: {}\n".format(value[0]+1, value[1][1]))

    # Carregar dades
    dades: pd.DataFrame = carregar(in_file)

    # Afegir noms columnes
    dades.columns: list = ["state", "city", "zipcode"]

    # Generar grups
    grouped: pd.DataFrameGroupBy = dades.groupby(["state", "city"])
    for name, group in grouped:
        create_folders(os.path.join(tgt_dir, name[0], name[1]))
        create_text_file(os.path.join(tgt_dir, name[0], name[1],
                                      "zipcode.txt"), group["zipcode"])


def print_directory_contents(tgt_dir: str):
    '''
    La funció imprimeix per pantalla lestructura
    de carpetes de primer nivell i el nombre de
    carpetes de segon nivell per a cada primer nivell.

    Aquesta funció accepta com a argument la ruta,
    en format string a partir de la qual realitzar l'analisi.

    La funció retorna per pantalla, la ruta de cada una
    de les carpetes (estats) que es troben en la ruta
    objectiu seguit del nombre de subcarpetes (ciutats)
    que hi ha dins de cada carpeta superior.
    '''
    # Generem un iterable amb el contingut de target_directory
    with os.scandir(tgt_dir) as dir_list:

        # Iterem a través de les entrades de contingut
        for entry in dir_list:

            # Quan el contingut es directory
            if entry.is_dir():

                # Generem un nou iterable amb el contingut del directory actual
                with os.scandir(os.path.join(
                     tgt_dir, entry.name, "")) as folder_list:

                    # Imprimim directory i nombre de carpetes
                    print("{}".format(os.path.join(tgt_dir, entry.name)))
                    print("{}".format(len(list(folder_list))))


# REFERENCIES
# https://pandas.pydata.org/docs/reference/groupby.html
# https://pandas.pydata.org/pandas-docs/stable/user_guide/groupby.html#splitting-an-object-into-groups
# https://stackoverflow.com/questions/30837089/how-to-create-directories-and-sub-directories-efficiently-and-elegantly-in-pytho
````

### Resposta 5.1

````python
# Resposta 5.1

import tarfile


def get_tree(path: str) -> list:
    '''
    La funció retorna el llistat d'arxius existents en
    la mateixa carpeta i totes les sub-carpetes a partir
    d'una ruta.

    Aquesta funció accepta com a argument la ruta, en
    format string a partir de la qual realitzar l'analisi.

    La funció retorna una llista amb la ruta de tots i cada
    un dels arxius.
    '''
    # Creem llista
    f_list: list = []

    # Iterem per tota la estructura de carpetes,
    # sub-carpetes i arxius a partir de la ruta d'entrada
    for dirpath, dirnames, filenames in os.walk(path):

        # Iterem només en els arxius trobats en la iteració anterior
        for one_file in filenames:

            # Afegim les rutes complertes dels arxius a la llista de retorn
            f_list.append(os.path.join(dirpath, one_file))
    return f_list


def tgz_folder(to_gz: str, dest: str):
    '''
    La funció comprimeix en un tar.gz tots els arxius
    existents en una carpeta i totes les seves sub-carpetes
    a partir d'una ruta.

    Aquesta funció accepta dos arguments:
    - to_gz: Ruta en format string de la carpeta pare
    que conté tots els arxius a comprimir. Els arxius
    poden estar en la mateixa carpeta o en sub-carpetes.
    - dest: Ruta i nom on generar l'arxiu comprimit tar.
    En el cas de rutes tipus: DIR/NOM, la funció generarà
    un arxiu tar anomenat NOM amb extensió .tar.gz, en la
    carpeta DIR.

    La funció retorna un arxiu comprimit tar incloent
    tots els arxius que s'han trobat en el directori
    i sub-directoris proporcionats.
    '''

    def comp_gz_list(input_list: list, gz_f_name: str):
        '''
        La funció genera un .tar amb tots els arxius
        especificats en la llista d'entrada.

        Aquesta funció accepta dos arguments:
        - input_list: llista amb totes les rutes dels
        arxius a comprimir en el tar.
        - gz_f_name: ubicació i nom de l'arxiu a generar

        La funció retorna un arxiu comprimit tar amb tots
        els arxius específicats en la llista d'entrada.
        '''
        # Generem la ruta per al arxiu tar
        cf: str = os.path.join(gz_f_name + ".tar.gz")

        # Obrim arxiu tar en mode escriptura
        with tarfile.open(cf, "w:gz") as tar:

            # Iterant per tots els arxius de la llista d'entrada,
            # afegim els arxius al tar.
            for to_gz_file in input_list:
                tar.add(to_gz_file)

    file_list: list = get_tree(to_gz)
    comp_gz_list(file_list, dest)
````

### Resposta 5.2

````python
# Resposta 5.2

def get_size(path: str) -> float:
    '''
    La funció obté el tamany de l'arxiu o directori
    proporcionat.

    Aquesta funció un argument de tipus str amb la ruta
    a l'arxiu o directori sobre el que calcular el tamany.

    La funció retorna en KB el tamany de l'arxiu o directori
    solicitat com a argument.
    '''

    # Inicialitzem tamany a zero
    size: float = 0

    # Comprovem si es arxiu
    if os.path.isfile(path):

        # Obtenim tamany de l'arxiu
        size = os.path.getsize(path)

    # Comprovem si es directory
    if os.path.isdir(path):

        # Obtenim llistat d'arxius en directori
        # i subdirectoris
        file_list: list = get_tree(path)

        # Iterem en tots els arxius
        for one_file in file_list:

            # Actualitzem el contador de tamany
            # afegint la mida de cada un dels arxius
            size += os.path.getsize(one_file)
    return size/1024

# REFERENCIES
# https://stackoverflow.com/questions/1392413/calculating-a-directorys-size-using-python
````
